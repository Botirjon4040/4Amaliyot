```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('Tabiat.jpg')
img = cv2.resize(img, (500, 300))

color_spaces = ('RGB','GRAY','HSV','LAB','XYZ','YUV')
color_image =[]
for color in color_spaces:

  if img is not None:
   converted_image = cv2.cvtColor(img, getattr(cv2, f'COLOR_BGR2'+color))
   color_image.append(converted_image)
  else:
   print("Error: Image 'Tabiat.jpg' could not be loaded.")

for i in range(len(color_spaces)):
 print(f"Displaying {color_spaces[i]} image:")
 cv2_imshow(color_image[i])
```
